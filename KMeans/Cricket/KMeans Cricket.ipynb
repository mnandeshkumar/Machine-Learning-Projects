```python
# Standardization
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_new)
```

```python
df_scaled
```

```python
df_df1 = pd.DataFrame(df_scaled,columns = col)
```

```python
df_df1
```

```python
# Start KMeans
```

```python
from sklearn.cluster import KMeans
```

```python
clusters = list(range(2,8))
ssd = []
for num_clusters in clusters:
    model_clus = KMeans(n_clusters = num_clusters,max_iter = 150, random_state = 42)
    model_clus.fit(df_df1)
    ssd.append(model_clus.inertia_)
```

```python
ssd
```

```python
plt.plot(clusters,ssd)
```

```python
kmodel = KMeans(n_clusters = 5, max_iter = 150, random_state = 42)
kmodel.fit(df_df1)
```

```python
kmodel.labels_
```

```python
df['ClusterID'] = kmodel.labels_
```
